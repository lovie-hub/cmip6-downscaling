```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Load files
ds1 = xr.open_dataset("AWI-OND3mm_ltm.nc")
ds2 = xr.open_dataset("AWI-NDJFMA3_ltm.nc")
ds3 = xr.open_dataset("AWI-MAM6_yearly.nc")

var = 'pr'  # replace with your variable name

# Extract data (adjust dimension name if needed)
data1 = ds1[var].mean(dim='time')
data2 = ds2[var].mean(dim='time')
data3 = ds3[var].mean(dim='time')

# Shared color scale
vmin = float(min(data1.min(), data2.min(), data3.min()))
vmax = float(max(data1.max(), data2.max(), data3.max()))

# Create figure
fig, axes = plt.subplots(1, 3, figsize=(18,6),
                         subplot_kw={'projection': ccrs.PlateCarree()})

datasets = [data1, data2, data3]
titles = ["OND Seasonal LTM ", "NDJFMA Seasonal LTM", "MAM Seasonal LTM"]

for ax, data, title in zip(axes, datasets, titles):
    im = data.plot(
        ax=ax,
        cmap='turbo',   # recommended rainfall colormap
        vmin=vmin,
        vmax=vmax,
        add_colorbar=False
    )

    # Map features
    ax.coastlines(resolution='50m')
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.LAND, facecolor='lightgray')
    ax.add_feature(cfeature.OCEAN, facecolor='white')

    # Gridlines
    gl = ax.gridlines(draw_labels=True, linestyle='--', alpha=0.5)
    gl.top_labels = False
    gl.right_labels = False

    ax.set_title(title)

# Single colorbar OUTSIDE
cbar_ax = fig.add_axes([0.55, 0.15, 0.02, 0.7])
cbar = fig.colorbar(im, cax=cbar_ax)
cbar.set_label('Precipitation (mm/day)')  # adjust units

# Adjust layout
plt.tight_layout(rect=[0, 0, 0.6, 1])
plt.show()
```

```
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from scipy.ndimage import gaussian_filter

# File paths
files = ["AWI-OND3mm_ltm.nc", "AWI-NDJFMA3_ltm.nc", "AWI-MAM6_yearly.nc"]
var = 'pr'  # replace with your variable name

data_list = []

# Load and smooth each dataset
for f in files:
    ds = xr.open_dataset(f)
    data = ds[var]
    # Average over time if it exists
    if 'time' in data.dims:
        data = data.mean(dim='time')
    # Gaussian smoothing (sigma=1)
    smoothed = xr.DataArray(
        gaussian_filter(data.values, sigma=1),
        coords=data.coords,
        dims=data.dims
    )
    data_list.append(smoothed)

# Shared color scale
vmin = float(min(d.min() for d in data_list))
vmax = float(max(d.max() for d in data_list))

# Create figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(18,6),
                         subplot_kw={'projection': ccrs.PlateCarree()})

titles = ["File 1", "File 2", "File 3"]

for ax, data, title in zip(axes, data_list, titles):
    im = ax.contourf(
        data['lon'], data['lat'], data,
        levels=20, cmap='YlGnBu', vmin=vmin, vmax=vmax
    )
    # Map features
    ax.coastlines(resolution='50m')
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.add_feature(cfeature.LAND, facecolor='lightgray')
    ax.add_feature(cfeature.OCEAN, facecolor='white')

    # Gridlines with labels
    gl = ax.gridlines(draw_labels=True, linestyle='--', alpha=0.5)
    gl.top_labels = False
    gl.right_labels = False

    ax.set_title(title, fontsize=12)

# Single colorbar outside
cbar_ax = fig.add_axes([0.92, 0.15, 0.02, 0.7])  # [left, bottom, width, height]
cbar = fig.colorbar(im, cax=cbar_ax)
cbar.set_label('mm/day', fontsize=12)

# Adjust layout
plt.tight_layout(rect=[0, 0, 0.9, 1])
plt.show()
```
